```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt

actualDatasetFull = pd.read_excel('CHEVRON_ACTUAL.xlsx')
targetDatasetFull = pd.read_excel('CHEVRON_TARGET.xlsx')
```

```python
uniqueCombinations = targetDatasetFull.drop_duplicates(subset=['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name'])

# Compare lengths
if len(targetDatasetFull) == len(uniqueCombinations):
    print("All records are unique based on the combination of the three columns.")
else:
    print("There are duplicate records based on the combination of the three columns.")
print(len(targetDatasetFull) , len(uniqueCombinations))
```

```python
duplicates = targetDatasetFull[['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name']][targetDatasetFull[['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name']].duplicated()]
duplicates
```

```python
duplicates = actualDatasetFull[['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name']][actualDatasetFull[['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name']].duplicated()]
duplicates
```

```python
actualDatasetFull = actualDatasetFull.dropna(subset = ['Plant Name'])
```

```python
groupedActual = actualDatasetFull.groupby(['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name','Item Type','Transit Type','Terminal Type','Refinery Area Desc','PW Source System','Refinery Area','Plant Status'])[['Volume BBL','Volume MBD','Book Volume BBL','Book Volume MBD']].sum().reset_index()groupedActual
```

```python
duplicates = groupedActual[['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name']][groupedActual[['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name']].duplicated()]
duplicates
```

```python
groupedTarget = targetDatasetFull.groupby(['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name','Target Type','PW Location Code','PW Location Level 0','PW Location Level 1','PW Location Level 2','PW Location Level 3'])[['Volume BBL','Volume MBD','Book Volume BBL','Book Volume MBD']].sum().reset_index()
```

```python
duplicates = groupedTarget[['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name']][groupedTarget[['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name']].duplicated()]
duplicates
```

```python
groupedActual = groupedActual.rename(columns={'Volume BBL':'Volume_BBL_Actual','Volume MBD':'Volume_MBD_Actual','Book Volume BBL':'Book_Volume_BBL_Actual','Book Volume MBD':'Book_Volume_MBD_Actual'})
groupedActual
```

```python
groupedTarget = groupedTarget.rename(columns={'Volume BBL':'Volume_BBL_Target','Volume MBD':'Volume_MBD_Target','Book Volume BBL':'Book_Volume_BBL_Target','Book Volume MBD':'Book_Volume_MBD_Target'})
groupedTarget
```

```python
merged_df = pd.merge(groupedTarget, groupedActual , on=['Stock Family', 'Stock Category New', 'Stock Category','Stock Sub Category','New Inventory Date','Plant Name'])
pd.set_option('display.max_columns', None)
merged_df
```

```python
# Check for missing values
missing_values = merged_df.isnull()
#print(missing_values)

# Count missing values in each column
missing_count = merged_df.isnull().sum()
print(missing_count)
```

```python
unique_values = merged_df.nunique()
print(unique_values)
```

```python
final_dataset = merged_df
```

```python
final_dataset['New Inventory Date'] = pd.to_datetime(final_dataset['New Inventory Date'])

# Extract year and month
final_dataset['Year'] = final_dataset['New Inventory Date'].dt.year
final_dataset['Month'] = final_dataset['New Inventory Date'].dt.strftime('%b')
final_dataset['Variance'] = ((final_dataset['Volume_MBD_Target']-final_dataset['Volume_MBD_Actual'])/final_dataset['Volume_MBD_Target'])*100
```

```python
final_dataset
```

```python
final_dataset.drop(['Item Type','Plant Status','Plant Name','New Inventory Date','PW Location Level 3','PW Location Level 2','PW Location Code','Stock Sub Category','Book_Volume_BBL_Target','Book_Volume_MBD_Target','Book_Volume_BBL_Actual','Book_Volume_MBD_Actual','Volume_BBL_Target','Volume_BBL_Actual'],axis=1, inplace=True)
```

```python
unique_values = final_dataset.nunique()
print(unique_values)
```
